```python
#visualize the CAFV Eligibility Simple (categorical data)
count=data['CAFV Eligibility Simple'].value_counts()
```

```python
count.plot(kind='bar',  grid= True)
plt.xlabel('CAFV Eligibility Simple')
plt.ylabel('Electric Vehicle Type')
plt.title('CAFV Eligibility Simple distribution')
plt.show()
```

```python
# which type of electric vehichle is eligible for CAFV Eligibility Simple

Eligibility_type=data.groupby('CAFV Eligibility Simple')['Electric Vehicle Type'].value_counts()
print(Eligibility_type)
```

here we find the exact number of types of vehicle which are eligible which are not

```python
Eligibility_type.plot(kind='bar',  grid= True, color='green')
plt.xlabel('CAFV Eligibility Simple')
plt.ylabel('distribution')
plt.title('CAFV Eligibility Simple distribution on based of type of EV')
plt.show()
```

by studying above data we find the relation between eligbility of CAFV and types of eletcric vehicle both data are categorical data which we need to plot into bar graph for better understanding âš–

**Acknowledgment of data: **

From the analysis of CAFV Eligibility for different vehicle types, we observe the following:

BEVs (Battery Electric Vehicles) are highly eligible under CAFV programs, with more than 40,000 vehicles marked as eligible.

Only a small number of BEVs (less than 10,000) are marked as non-eligible, showing that most BEVs meet the environmental and efficiency standards required for eligibility.

PHEVs (Plug-in Hybrid Electric Vehicles) have a mixed eligibility status: around 22,000 are eligible, but a similarly large number (20,000+) are non-eligible.

This suggests that not all PHEVs meet the CAFV requirements, possibly due to lower electric range or higher emissions compared to BEVs.

âœ… Final Insight:
BEVs are more consistently eligible for clean vehicle programs like CAFV due to their fully electric nature and zero emissions, whereas PHEVs show variability in eligibility based on their hybrid design and performance.






```python
# which car maker provide how many electric range
plt.hist(data['Electric Range'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.title('Distribution of Electric Range')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

ðŸ§  Final Conclusion: The histogram of electric range shows a right-skewed distribution, indicating that most electric vehicles offer lower to moderate range, typically below 150 miles, while a few models provide extended ranges above 250 miles. This suggests a market trend where affordable EVs prioritize lower range, while premium models invest in longer driving capabilities.
